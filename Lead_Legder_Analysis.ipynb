```python
import pandas as pd

leads = pd.read_excel("Lead_ledger_sample.xlsx")
leads.head()
```

```python
# Let's create a copy containing only our columns of focus

leads_copy = leads[['Company','Title','Location','State']]
```

```python
# Let's quicky check the dataframe

leads_copy
```

```python
# Calculating the count of leads by State

pd.crosstab(index=leads_copy['State'], columns='Num_Of_Companies', 
            values=leads_copy['Company'], aggfunc='count').sort_values(by='Num_Of_Companies', ascending=False)
```

```python
# Calculate the count of leads by Location

pd.crosstab(index=leads_copy['Location'], columns='Num_Of_Companies', 
            values=leads_copy['Company'], aggfunc='count').sort_values(by='Num_Of_Companies', ascending=False)
```

```python
# Calculating the count of leads by State

pd.crosstab(index=leads_copy['Title'], columns='Num_Of_Companies', 
            values=leads_copy['Company'], aggfunc='count').sort_values(by='Num_Of_Companies', ascending=False)
```

### Insights 

**Clearly, as per the small sample of data that we have;** 

  * California, New York, Illinois are the hot zone segments that overindex.
  * Within these states San Francisco Bay Area, Greater New York City Area, Los Angeles Metropolitan Area, etc. are zoomed-in hot zone segments.
  * Owner and President Titles are better leads.
  
**Please note that the above insights could change once we have the complete data**
